```python
import pandas as pd
```

```python
df = pd.read_csv("devnagari.csv")
```

```python
df.shape #46 Characters with each 2000 samples
```

```python
df.columns
```

```python
selected_chars = ['character_04_gha', 'character_14_dhaa',
                  'character_24_bha','character_25_ma',
                  'character_21_pa','character_22_pha']
```

```python
df_s = df.loc[df['character'].isin(selected_chars), :]
```

```python
df_s.shape
```

```python
x = df_s.drop('character',axis=1)
y = df_s.loc[:,'character']
```

```python
# label encode y column
label_map = {'character_04_gha':0, 'character_14_dhaa':1, 'character_21_pa':2,
       'character_22_pha':3, 'character_24_bha':4, 'character_25_ma':5}
y = y.map(label_map)
```

```python
y.value_counts()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,
                                                 test_size=0.3,
                                                 random_state=7,
                                                 stratify=y)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import classification_report
```

```python
l = {"Logistic":LogisticRegression(random_state=117,max_iter=1000,class_weight="balanced"),
     "GaussianNB":GaussianNB(),"BernoulliNB":BernoulliNB(),"Multinomial":MultinomialNB(),
     "svc":SVC(random_state=7,kernel='rbf'),
     "KNN":KNeighborsClassifier(),
     "DT":DecisionTreeClassifier(class_weight="balanced")}
```

```python
for k,v in l.items():
    model = v
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    print("\nClassification Report:",k)
    print(classification_report(y_test,y_pred))
```

```python
#Random_Forest_Classifier
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(max_samples=0.8,random_state=117,oob_score=True)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
#XGBoost
from xgboost import XGBClassifier
model = XGBClassifier()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
#CATBoost
from catboost import CatBoostClassifier
model = CatBoostClassifier()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print(classification_report(y_test,y_pred))
```
