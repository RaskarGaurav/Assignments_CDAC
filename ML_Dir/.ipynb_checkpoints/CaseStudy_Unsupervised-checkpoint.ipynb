```python
#Knee plot
from sklearn.neighbors import NearestNeighbors
from matplotlib import pyplot as plt
neigh = NearestNeighbors(n_neighbors=2)
nbrs = neigh.fit(df)
distances, indices = nbrs.kneighbors(df)
distances = np.sort(distances, axis=0)
distances = distances[:,1]
fig, ax = plt.subplots()
ax.plot(distances)
ax.minorticks_on()
ax.grid(True, which='minor')
```

```python
#OUTLIERS...
```

```python
#After Standard Scaling 99% data is between +3 & -3
#So value less than -3 and greater than +3 are considered as outliers
#Impute the values
def out(df1,col):
    df1.loc[df1[col]>3,col]=3
    df1.loc[df1[col]<-3,col]=-3
    return df1

for i in df1.columns:
    df1 = out(df1,i)
```

```python
df1.head()
```

```python
df1.describe()
```

```python
#IQR outliers imputing
def IQROut(df,col):
    q1,q3=df[col].quantile([0.25,0.75])
    iqr = q3-q1
    minRange = q1 - 1.5*iqr
    maxRange = q3 + 1.5*iqr

    df.loc[df[col]<minRange,col] = minRange
    df.loc[df[col]>maxRange,col] = maxRange
    return df
l = ("carat","depth","table","price","x","y","z")
for i in l:
    x_train = IQROut(x_train,i)
```

```python
x_train.head()
```

```python
#KMeans
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
inertia = []
for i in range(2,11):
    km = KMeans(n_clusters=i,random_state=117,max_iter=100)
    km.fit(x_train)
    inertia.append(km.inertia_)
```

```python
plt.plot(range(2,11),inertia,marker ="o")
```

```python
model = KMeans(n_clusters=4,max_iter=100,random_state=117)
model.fit(x_train)
```

```python
x_pred = model.predict(x_test)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x_test,x_pred)
```
