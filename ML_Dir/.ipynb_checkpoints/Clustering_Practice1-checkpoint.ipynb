```python
#KMeans
```

```python
from sklearn.cluster import KMeans
```

```python
Mk = KMeans(n_clusters=3,max_iter=300,random_state=117)
```

```python
y_pred = Mk.predict(x_test)
```

```python
#Dendrogram
import matplotlib.pyplot as plt

import scipy.cluster.hierarchy as shc
plt.figure(figsize=(5, 3))
plt.title("Dendrograms")
dend = shc.dendrogram(shc.linkage(df.iloc[:500,:], method='ward'))
plt.axhline(y=50, color='r', linestyle='--')
```

```python
#DBSCAN
from sklearn.cluster import DBSCAN
```

```python
#Standard Scaler
x_train_str = (x_train- x_train.mean())/x_train.std()
```

```python
x_train_str.std(),x_train_str.mean()
```

```python
DBscan = DBSCAN(eps= 0.2)
```

```python
DBscan.fit(x_train_str.iloc[:500])
```

```python
DBscan.fit_predict(x_train_str.iloc[:500])
```
