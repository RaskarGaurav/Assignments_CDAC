```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("BankChurners_proper.csv",na_values="Unknown")
```

```python
df.shape
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.dtypes
```

```python
df.head()
```

```python
df["Attrition_Flag"].unique()
```

```python
df["Education_Level"].unique()
```

```python
labels = {"Attrition_Flag":{'Existing Customer':1, 'Attrited Customer':0}}
```

```python
df.replace(labels,inplace=True)
```

```python
df.head()
```

```python
l=df.select_dtypes(include="object").columns
```

```python
df = pd.get_dummies(df,columns=l)
```

```python
df.shape
```

```python
df.head()
```

```python
x= df.drop("Attrition_Flag",axis=1)
y= df.iloc[:,0]
```

```python
y.shape
```

```python
#Standard Scaling
x = (x-x.mean())/x.std()
```

```python
x.describe()  
```

```python
#outliers
def out(df,col):
    df.loc[df[col]>3,col] = 3
    df.loc[df[col]<-3,col] = -3
    return df

for i in x.columns:
    df = out(x,i)
```

```python
#Train_Test_Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=117,test_size=0.3)
```

```python
x_train.shape , y_train.shape
```

```python
#PCA
from sklearn.decomposition import PCA
```

```python
pca1 = PCA(random_state=117)
```

```python
pca1.fit(x_train)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(x.shape[1]),pca1.explained_variance_ratio_)
plt.grid(True)
plt.minorticks_on()
```

```python
pcaModel = PCA(n_components=0.95,random_state=117)
```

```python
pcaModel.fit(x_train)
```

```python
x_train_pca = pcaModel.transform(x_train)
x_test_pca = pcaModel.transform(x_test)
```

```python
x_train_pca = pd.DataFrame(x_train_pca,index = x_train.index)
x_test_pca = pd.DataFrame(x_test_pca,index = x_test.index)
```

```python
x_train_pca.shape , x_test_pca.shape
```

```python
x_train_pca.head()
```

```python
x_test_pca.head()
```

```python
y.value_counts()
```

```python
#Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
Model = DecisionTreeClassifier(class_weight="balanced",random_state=117)
```

```python
Model.fit(x_train_pca,y_train)
```

```python
y_pred = Model.predict(x_test_pca)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print(100*accuracy_score(y_test,y_pred))
```

```python
#Select Import Features
from sklearn.feature_selection import SelectFromModel
```

```python
dt = DecisionTreeClassifier(class_weight="balanced",random_state=117)
sfm = SelectFromModel(dt,threshold="mean")
```

```python
sfm.fit(x_train,y_train)
```

```python
x_train_dt = sfm.transform(x_train)
x_test_dt = sfm.transform(x_test)
```

```python
x_train_dt.shape , x_test_dt.shape
```

```python
dt.fit(x_train_dt,y_train)
```

```python
y_pred_dt = dt.predict(x_test_dt)
```

```python
print(100*accuracy_score(y_test,y_pred_dt))
```
