```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("cleaned_76hd.csv")
```

```python
df.shape
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop("V58",axis=1)
y = df['V58']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=117,stratify=y)
```

```python
#Decison Tree Classifier
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
modelDT = DecisionTreeClassifier(class_weight="balanced")
```

```python
modelDT.fit(x_train,y_train)
y_predDT = modelDT.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_recall_fscore_support,confusion_matrix,classification_report
100*accuracy_score(y_test,y_predDT)
```

```python
print(classification_report(y_test,y_predDT))
```

```python
confusion_matrix(y_test,y_predDT)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
modelLR = LogisticRegression(random_state=117,class_weight="balanced",solver="saga",max_iter=300)
```

```python
modelLR.fit(x_train,y_train)
```

```python
y_predLR = modelLR.predict(x_test)
```

```python
100*accuracy_score(y_test,y_predLR)
```

```python
print(classification_report(y_test,y_predLR))
```

```python
#Naive Bayes
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
```

```python
modelGB = GaussianNB()
modelBB = BernoulliNB()
modelMB = MultinomialNB()
```

```python
modelGB.fit(x_train,y_train)
modelBB.fit(x_train,y_train)
modelMB.fit(x_train,y_train)
```

```python
y_predGB = modelGB.predict(x_test)
y_predBB = modelBB.predict(x_test)
y_predMB = modelMB.predict(x_test)
```

```python
print(100*accuracy_score(y_test,y_predGB))
print(100*accuracy_score(y_test,y_predBB))
print(100*accuracy_score(y_test,y_predMB))
```

```python
#SVM
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(random_state=7,kernel='rbf',gamma="auto",C=1)
svc.fit(x_train,y_train)
y_pred = svc.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
#KNN
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
modelKNN = KNeighborsClassifier()
modelKNN.fit(x_train,y_train)
y_predNN = modelKNN.predict(x_test)
print(classification_report(y_test,y_predNN))
```
