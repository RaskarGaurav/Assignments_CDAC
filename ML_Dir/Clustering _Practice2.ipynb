```python
df = pd.get_dummies(df,columns=["cut"])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test = train_test_split(df,test_size=0.3,random_state=117)
```

```python
x_train.shape ,x_test.shape
```

```python
#KMeans
from sklearn.cluster import KMeans
```

```python
mk = KMeans(n_clusters=3,max_iter=100,random_state=117)
```

```python
mk.fit(x_train)
```

```python
x_pred = mk.predict(x_test)
```

```python
#Dendrogram
import matplotlib.pyplot as plt

import scipy.cluster.hierarchy as shc
plt.figure(figsize=(5, 3))
plt.title("Dendrograms")
dend = shc.dendrogram(shc.linkage(df.iloc[:500,:], method='ward'))
plt.axhline(y=50, color='r', linestyle='--')
```

```python
#ELBOW Plot
inertia =[]
for k in range(2,11):
    km = KMeans(n_clusters=k,random_state=117)
    km.fit(x_train)
    inertia.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(2,11),inertia)
```
